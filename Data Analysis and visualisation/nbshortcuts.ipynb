```python
import random
import pandas as pd

customer_names = ["John Travolta","Judie Dench","Lewis Capaldi","Bruce Forsyth","Andy Goode","James Haskel"]

customer_purchases = ["mouse", "keyboard","dongle","Laptop","MiniPC","5090"]

expanded_data = {
    "Customer Name":[random.choice(customer_names) for _ in range(100)],
    "Purchase Items":[", ".join(random.sample(customer_purchases, random.randint(1,3)))
    for i in range(100)],
    "Purchase Costs":[round(random.uniform(50,1500),2) for _ in range(100)]
    }

# print(len(expanded_data["Customer Name"]))
# print(len(expanded_data["Purchase Items"]))
# print(len(expanded_data["Purchase Costs"]))

customer_dataset = pd.DataFrame(expanded_data)

customer_dataset
```

```python
grouped_dataset = customer_dataset.groupby("Customer Name").sum().reset_index()

grouped_dataset
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

#plot bar chart
plt.figure(figsize=(15,5))
plt.bar(x=grouped_dataset["Customer Name"],height=grouped_dataset["Purchase Costs"],data=grouped_dataset)
```
